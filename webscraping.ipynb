```python
from bs4 import BeautifulSoup

# Path to your saved HTML file
file_path = "C:\\Users\\Player One\\Desktop\\Weinfrst.txt"

with open(file_path, 'r', encoding='utf-8') as f:
    html = f.read()

soup = BeautifulSoup(html, 'html.parser')

# Find all rating divs
rating_divs = soup.find_all('div', class_='ruk_rating_snippet')

for div in rating_divs:
    aria_label = div.get('aria-label', '')
    if aria_label:
        parts = aria_label.split(' ')
        rating = parts[0]        # e.g. "4.3"
        num_ratings = parts[-1]  # e.g. "236"
        print(f"Rating: {rating}, Number of ratings: {num_ratings}")
```

```python
ratings = []
num_ratings = []

for div in rating_divs:
    aria_label = div.get('aria-label', '')
    if aria_label:
        parts = aria_label.split(' ')
        ratings.append(parts[0])
        num_ratings.append(parts[-1])
    else:
        ratings.append("0")      # or 0 if you want integer
        num_ratings.append("0")  # same here

print(ratings)
print(num_ratings)
```

```python
len(ratings)
```

```python
len(num_ratings)
```

```python
products = soup.find_all('div', class_='product-item-main')
print(f'Found {len(products)} products')
```

```python
main_prices = []

for product in products:
    # Within each product, find price spans without discount class
    price_span = product.find('span', class_='price')
    if price_span:
        classes = price_span.get('class', [])
        if 'price--highlight' in classes:
            # find non-discount price inside the same product
            non_discount_price = product.find('span', class_=lambda c: c and 'price' in c and 'price--highlight' not in c)
            if non_discount_price:
                main_prices.append(non_discount_price.get_text(strip=True))
            else:
                main_prices.append(price_span.get_text(strip=True))
        else:
            main_prices.append(price_span.get_text(strip=True))
    else:
        main_prices.append('N/A')

print(len(main_prices))
print(main_prices)
```

```python
#create database

import pandas as pd

df = pd.DataFrame({
    'Producer': producers,
    'Product': products,
    'Price': main_prices,
    'Rating': ratings,
    'Number of Ratings': num_ratings
})

print(df)
```
