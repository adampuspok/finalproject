CLUSTERING

```python
# k-means clustering

from sklearn.preprocessing import StandardScaler

cluster_vars = ['income', 'age', 'tenure', 'total_spent', 'family_size', 'is_parent']
X = df[cluster_vars].copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Define range of k to test
k_range = range(1, 11)


inertia = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(8, 4))
plt.plot(list(k_range), inertia, marker='o')  # Make sure both x and y are same length
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Apply KMeans with k=4
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

# Choose min_samples you used (e.g., 5)
min_samples = 5

# Fit nearest neighbors model on scaled data
nbrs = NearestNeighbors(n_neighbors=min_samples)
nbrs.fit(X_scaled)

# Find distances to k-th nearest neighbors for each point
distances, indices = nbrs.kneighbors(X_scaled)

# Take distances to the k-th neighbor for each point (last column)
k_distances = distances[:, -1]

# Sort distances ascending
k_distances = np.sort(k_distances)

# Plot the sorted k-distances
plt.figure(figsize=(8,5))
plt.plot(k_distances)
plt.xlabel('Points sorted by distance')
plt.ylabel(f'{min_samples}-th nearest neighbor distance')
plt.title('k-distance Graph to choose eps for DBSCAN')
plt.grid(True)
plt.show()
```

```python
for eps_val in [1.0, 1.1, 1.2]:
    dbscan = DBSCAN(eps=eps_val, min_samples=5)
    labels = dbscan.fit_predict(X_scaled)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)
    print(f"eps={eps_val:.1f} --> clusters: {n_clusters}, noise points: {n_noise}")
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Example: K-Means with k=6
kmeans = KMeans(n_clusters=6, random_state=42).fit(X_scaled)
labels_kmeans = kmeans.labels_

# DBSCAN with eps=1.2 and min_samples=5
dbscan = DBSCAN(eps=1.2, min_samples=5).fit(X_scaled)
labels_dbscan = dbscan.labels_

# Calculate metrics (ignore noise for DBSCAN in silhouette)
mask = labels_dbscan != -1

print("K-Means:")
print(" Silhouette:", silhouette_score(X_scaled, labels_kmeans))
print(" Davies-Bouldin:", davies_bouldin_score(X_scaled, labels_kmeans))
print(" Calinski-Harabasz:", calinski_harabasz_score(X_scaled, labels_kmeans))

print("\nDBSCAN:")
if sum(mask) > 0:
    print(" Silhouette:", silhouette_score(X_scaled[mask], labels_dbscan[mask]))
    print(" Davies-Bouldin:", davies_bouldin_score(X_scaled[mask], labels_dbscan[mask]))
    print(" Calinski-Harabasz:", calinski_harabasz_score(X_scaled[mask], labels_dbscan[mask]))
else:
    print(" No valid clusters for DBSCAN")
```

```python
# As data has low noise and K-Means performs better on cluster validity metrics, I will proceed with K-Means
```

HYPOTHESIS TESTING

```python
import pandas as pd
df = pd.read_csv('marketingc_clean.csv')
```

```python
# Hypothesis 1: Do customers in different clusters show different levels of recency?
# H0: There is no significant difference in recency across clusters.
# H1: There is a significant difference in recency across clusters.

from scipy.stats import f_oneway

grouped_recency = [group['recency'].values for name, group in df.groupby('cluster')]

f_stat, p_val = f_oneway(*grouped_recency)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_val:.5f}")

# Interpretation
if p_val < 0.05:
    print("Significant differences found in recency across clusters (reject H₀).")
else:
    print("No significant difference in recency across clusters (fail to reject H₀).")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='cluster', y='recency', palette='pastel')

plt.title('Recency Distribution by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Recency (Days Since Last Purchase)')
plt.tight_layout()
plt.show()
```

```python
# Hypothesis 2: Do customers in different clusters respond differently?
# H0: Response rates are the same across all clusters
# H1: Response rates differ across clusters

import pandas as pd
from scipy.stats import chi2_contingency

# Create cross-tab (contingency table)
contingency = pd.crosstab(df['cluster'], df['response'])

print(contingency)
```

```python
chi2, p, dof, expected = chi2_contingency(contingency)

print(f"Chi-square statistic: {chi2:.3f}")
print(f"P-value: {p:.5f}")

if p < 0.05:
    print("Significant difference in response rates across clusters (reject H₀).")
else:
    print("No significant difference in response rates (fail to reject H₀).")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate response rate per cluster
response_rates = df.groupby('cluster')['response'].mean().reset_index()

sns.barplot(data=response_rates, x='cluster', y='response')
plt.title('Response Rate by Cluster')
plt.ylabel('Response Rate')
plt.xlabel('Cluster')
plt.ylim(0, 1)
plt.show()
```

```python
# Hypothesis 3: Do acceptance rates differ significantly between campaigns?
# H0: Acceptance rates are the same across all campaigns
# H1: Acceptance rates differ across campaigns

from statsmodels.stats.contingency_tables import cochrans_q
import pandas as pd

cmp_cols = ['acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3', 'acceptedcmp4', 'acceptedcmp5']
campaign_data = df[cmp_cols]

result = cochrans_q(campaign_data)

print(f"Cochran's Q statistic: {result.statistic:.3f}")
print(f"P-value: {result.pvalue:.5f}")

if result.pvalue < 0.05:
    print("➡️ Significant differences between campaigns' acceptance rates.")
else:
    print("➡️ No significant difference found.")
```

```python
acceptance_rates = df[cmp_cols].mean()

cmp_cols_ordered = ['acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3', 'acceptedcmp4', 'acceptedcmp5']

acceptance_rates = acceptance_rates.reindex(cmp_cols_ordered)

acceptance_rates.plot(kind='barh', color='skyblue')
plt.title('Campaign Acceptance Rates')
plt.xlabel('Acceptance Rate')
plt.tight_layout()
plt.show()
```

```python
# Hypothesis 3: Do customers in different clusters have different average incomes?

from scipy.stats import spearmanr

corr_coef, p_value = spearmanr(df['age'], df['total_spent'])

print(f"Spearman correlation coefficient: {corr_coef:.3f}")
print(f"P-value: {p_value:.5f}")

if p_value < 0.05:
    print("Significant correlation between age and total_spent.")
else:
    print("No significant correlation found.")
```
